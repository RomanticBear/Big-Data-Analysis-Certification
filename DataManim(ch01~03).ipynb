```python
# 20번 데이터를 로드하라.

file_path = "https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv"
df=pd.read_csv(file_path)

df.head()
```

```python
# 21번 quantity컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하라

ans_col=df[df['quantity']==3]
ans_col.head(5)
```

```python
# 22번 quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하라

ans=df[df['quantity']==3].reset_index(drop=True)

ans.head(5)
```

```python
# 23번 quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라

ans=df[['quantity','item_price']]

ans
```

```python
# 24번 item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라

df['new_price']=df["item_price"].str[1:].astype('float')
ans.head()
```

```python
# 25번 new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라

ans=len(df[df['new_price']<=5])

ans
```

```python
# 26번 item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하라고 index 값을 초기화 하여라


ans=df[df.item_name=='Chicken Salad Bowl'].reset_index(drop=True)

ans.head()
```

```python
# 27번 new_price값이 9 이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임을 추출하라


Ans = df.loc[(df.item_name =='Chicken Salad Bowl') & (df.new_price <= 9)]

Ans.head(4)
```

```python
df.sort_values('new_price').reset_index(drop=True)

df.head()
```

```python
# 29번 df의 item_name 컬럼 값중 Chips 포함하는 경우의 데이터를 출력하라

Ans=df[df['item_name'].str.contains('Chips')]

Ans.head()
```

```python
# 30번 df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라

df.iloc[:,::2]
```

```python
# 31번 df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하여라

ans=df.sort_values('new_price',ascending=False).reset_index()

ans
```

```python
# 32 df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱싱하라

df[(df['item_name']=="Steak Salad")|(df['item_name']=="Bowl")]
```

```python
# 33번 df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

ans=df[(df['item_name']=="Steak Salad")|(df['item_name']=="Bowl")]
ans=ans.drop_duplicates('item_name')

ans
```

```python
# 34번 df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라

Ans = df.loc[(df.item_name =='Steak Salad') | (df.item_name =='Bowl')]
Ans = Ans.drop_duplicates('item_name',keep='last')

# Ans
print(help(pd.DataFrame.drop_duplicates))
```

```python
# 35번 df의 데이터 중 new_price값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱하라

ans=df[df['new_price']>=df['new_price'].mean()]

ans
```

```python
# 36번 df의 데이터 중 item_name의 값이 Izze 데이터를 Fizzy Lizzy로 수정하라 

df.loc[df['item_name']=='Izze','item_name']='Fizzy Lizzy'

df
```

```python
# 37번 df의 데이터 중 choice_description 값이 NaN 인 데이터의 갯수를 구하여라

# print(dir(pd.DataFrame))

ans=df['choice_description'].isnull().sum()

ans
```

```python
# 39번 df의 데이터 중 choice_description 값에 Black이 들어가는 경우를 인덱싱하라


ans=df[df['choice_description'].str.contains('Black')]

ans.head()
```

# 03_Grouping


```python
# 44번 데이터를 로드하고 상위 5개 컬럼을 출력하라

df=pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv")

df.head()
```

```python
# 45번 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라

# 방법 1
ans=df.groupby("host_name").size()


# 방법 2
ans=df["host_name"].value_counts().sort_index()
```

```python
# 46번 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라


ans=df.groupby("host_name").size().to_frame().rename(columns={0:'counts'})

ans2=ans.sort_values('counts',ascending=False)

ans2.head()
```
