```python

# 56Î≤à 

df=pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv")


# 57Î≤à Income_CategoryÏùò Ïπ¥ÌÖåÍ≥†Î¶¨Î•º map Ìï®ÏàòÎ•º Ïù¥Ïö©ÌïòÏó¨ Îã§ÏùåÍ≥º Í∞ôÏù¥ Î≥ÄÍ≤ΩÌïòÏó¨ newIncome Ïª¨ÎüºÏóê Îß§ÌïëÌïòÎùº

df
dic={
    'Unknown': 'N',
    'Less than $40K': 'a',
    '$40K - $60K': 'b',
    '$60K - $80K': 'c',
    '$80K - $120K': 'd',
    '$120K +': 'e'
}

df['newIncome']=df["Income_Category"].map(lambda x:dic[x])

ANS=df['newIncome']

ANS.head()


```

```python
# 58Î≤à Income_CategoryÏùò Ïπ¥ÌÖåÍ≥†Î¶¨Î•º apply Ìï®ÏàòÎ•º Ïù¥Ïö©ÌïòÏó¨ Îã§ÏùåÍ≥º Í∞ôÏù¥ Î≥ÄÍ≤ΩÌïòÏó¨ newIncome Ïª¨ÎüºÏóê Îß§ÌïëÌïòÎùº

def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'


df['newIncome']=df['Income_Category'].apply(changeCategory)

ans=df['newIncome']

ans.head()
```

```python
# 59Î≤à Customer_AgeÏùò Í∞íÏùÑ Ïù¥Ïö©ÌïòÏó¨ ÎÇòÏù¥ Íµ¨Í∞ÑÏùÑ AgeState Ïª¨ÎüºÏúºÎ°ú Ï†ïÏùòÌïòÎùº. (0~9 : 0 , 10~19 :10 , 20~29 :20 ‚Ä¶ Í∞Å Íµ¨Í∞ÑÏùò ÎπàÎèÑÏàòÎ•º Ï∂úÎ†•ÌïòÎùº

df['AgeState']=df['Customer_Age'].map(lambda x:x//10*10)

Ans=df['AgeState'].value_counts().sort_index()

Ans
```

```python
# 60Î≤à Education_LevelÏùò Í∞íÏ§ë GraduateÎã®Ïñ¥Í∞Ä Ìè¨Ìï®ÎêòÎäî Í∞íÏùÄ 1 Í∑∏Î†áÏßÄ ÏïäÏùÄ Í≤ΩÏö∞ÏóêÎäî 0ÏúºÎ°ú Î≥ÄÍ≤ΩÌïòÏó¨ newEduLevel Ïª¨ÎüºÏùÑ Ï†ïÏùòÌïòÍ≥† ÎπàÎèÑÏàòÎ•º Ï∂úÎ†•ÌïòÎùº


def func(x):
    if 'Graduate' in x :
        return 1
    else:
        return 0

df["newEduLevel"]=df["Education_Level"].apply(func)

ans = df['newEduLevel'].value_counts()

ans
```

```python
# 61Î≤à Credit_Limit Ïª¨ÎüºÍ∞íÏù¥ 4500 Ïù¥ÏÉÅÏù∏ Í≤ΩÏö∞ 1 Í∑∏Ïô∏Ïùò Í≤ΩÏö∞ÏóêÎäî Î™®Îëê 0ÏúºÎ°ú ÌïòÎäî newLimit Ï†ïÏùòÌïòÎùº. newLimit Í∞Å Í∞íÎì§Ïùò ÎπàÎèÑÏàòÎ•º Ï∂úÎ†•ÌïòÎùº

df.head()

df['newLimit']=df['Credit_Limit'].map(lambda x:1 if x>=4500 else 0)


ANS=df['newLimit'].value_counts()

ANS
```

```python
# 62Î≤à 

def check(x):
    if x['Marital_Status']=='Married' and x['Card_Category']=='Platinum':
        return 1
    else:
        return 0


df['newState']=df.apply(check,axis=1)

ANS=df['newState'].value_counts()

ANS
```

```python
# 63Î≤à - ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ ÏÜêÏÉÅ ÏãúÌÇ®ÎìØ ,,??

def check(x):
    if x=='M':
        return 'male'
    else:
        return 'female'

df['Gender']=df['Gender'].apply(check)

ans=df['Gender'].value_counts()

ans
```
